# 2. Datenbereinigung und -aufbereitung
Der Datensatz wird auf die für die Fragestellung relevanten Unfalldaten des Bundeslandes Hamburg reduziert. Zudem werden alle Unfalldaten entfernt, die keine Radbeteiligung aufweisen.

```python
# Aus dem Datensatz werden alle Unfalldaten entfernt, die nicht das Bundesland Hamburg (Wert = 2) betreffen. Gleichzeitig werden alle Unfalldaten entfernt, an denen keine Radfahrer beteiligt waren. 
df_hh_rad = df_concat.loc[(df_concat['ULAND'] == 2) & (df_concat['IstRad']==1)]
```

```python
df_hh_rad.head()
```

```python
# Kontrolle, wie viele Unfalldaten mit Radbeteiligung im Bundesland Hamburg in den Jahre 2016-2019 vorliegen
df_hh_rad["IstRad"].value_counts()
```

```python
# Nutzung des einheitlichen Datensatzes der Gruppe (merged_csv)
file_path_mergeddata = "data/merged_csv.csv"
df_hh_rad = pd.read_csv(file_path_mergeddata, delimiter=';',decimal=',')
```

```python
df_hh_rad.info()
# To do: Warum taucht hier wieder die Spalte FID auf?
```

```python
# Umwandlung der Kategorie "UWOCHENTAG" von Zahlenwerten in Worte
df_hh_rad.loc[df_hh_rad['UWOCHENTAG']==1, "UWOCHENTAG"]= "Sonntag"
df_hh_rad.loc[df_hh_rad['UWOCHENTAG']==2, "UWOCHENTAG"]= "Montag"
df_hh_rad.loc[df_hh_rad['UWOCHENTAG']==3, "UWOCHENTAG"]= "Dienstag"
df_hh_rad.loc[df_hh_rad['UWOCHENTAG']==4, "UWOCHENTAG"]= "Mittwoch"
df_hh_rad.loc[df_hh_rad['UWOCHENTAG']==5, "UWOCHENTAG"]= "Donnerstag"
df_hh_rad.loc[df_hh_rad['UWOCHENTAG']==6, "UWOCHENTAG"]= "Freitag"
df_hh_rad.loc[df_hh_rad['UWOCHENTAG']==7, "UWOCHENTAG"]= "Samstag"
```

```python
df_hh_rad.loc[df_hh_rad['ULICHTVERH']==0, "ULICHTVERH"]= "Tageslicht"
df_hh_rad.loc[df_hh_rad['ULICHTVERH']==1, "ULICHTVERH"]= "Dämmerung"
df_hh_rad.loc[df_hh_rad['ULICHTVERH']==2, "ULICHTVERH"]= "Dunkelheit"
## To do: Warum funktioniert Umbenennung nicht? (Object?)
```
